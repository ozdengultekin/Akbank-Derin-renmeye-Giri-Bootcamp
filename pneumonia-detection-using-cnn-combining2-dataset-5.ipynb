# VER N LEME ve GRSELLETRME

**Snflar ve Grnt Boyutu:**<br>
Labels listesi, veri setindeki snflar temsil eder: PNEUMONIA ve NORMAL. Tm grntler 150x150 piksel boyutuna getirilir; bu, CNN modellerinde giri boyutunun sabit olmas iin gereklidir.<br>
**n leme:**
**CLAHE (Contrast Limited Adaptive Histogram Equalization)** ile grnt kontrast artrlr. Bylece akcier dokusundaki kk detaylar ve opasiteler daha belirgin hle gelir. Normalizasyon uygulanrsa, piksel deerleri 01 aralna leklenir; bu, modelin daha hzl ve kararl renmesini salar. Son olarak reshape ilemi ile grntler CNN giri formatna (ykseklik, genilik, kanal) getirilir.<br>
**Veri Setinin Hazrlanmas:**
Her resim, uygun preprocessing admlarndan getikten sonra kendi snf etiketi ile birlikte [resim, etiket] ifti hline getirilir. Bu sayede veri seti modelin eitimine hazr hale gelir.<br>
**zellikler ve Etiketlerin Ayrlmas:**
X dizisi modelin giriine verilecek grntleri, Y dizisi ise her grntnn snf etiketini ierir. Bu ayrm, modelin giri ve hedef dizilerini net bir ekilde ayrr.

````python
labels = ["PNEUMONIA", "NORMAL"]
img_size = 150  # 150x150 piksel

#Normalizasyon ve Yeniden Boyutlandrma

def preprocess_image(img):
    # CLAHE ile kontrast artr
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img = clahe.apply(img)
    # Normalize et
    img = img.astype(np.float32) / 255.0
    # CNN iin reshape
    img = img.reshape(img_size, img_size, 1)
    return img

def get_training_data(data_dir):
    dataset = []
    for class_id, label in enumerate(labels):
        label_dir = os.path.join(data_dir, label)
        for file in os.listdir(label_dir):
            img_path = os.path.join(label_dir, file)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = cv2.resize(img, (img_size, img_size))
                img = preprocess_image(img)  # n ileme + normalize + reshape
                dataset.append([img, class_id])
    return np.array(dataset)

# zellikler ve etiketleri ayr
def split_features_labels(dataset):
    x = [feature for feature, label in dataset]
    y = [label for feature, label in dataset]
    return np.array(x), np.array(y)
````

**PNEUMONIA** ve **NORMAL** klasrnde bulunan resimleri:

* Grayscale (siyah-beyaz) olarak okur,150x150 boyutuna kltr. img_size klasrlerde bulunan her resim boyutu farkl. Bu grntleri ilerken her bir grntnn eit boyutta olmas iin bu boyutu her resim iin ayn boyuta getiriryoruz. Bu ayrca modelin resimleri daha hzl ilemesini salar. Resim bydke modelin eitim sresinin uzad da gzlemlenmitir.
PNEUMONIA 0 , NORMAL 1 olarak listede tanmland.
* Her resmi ilgili snf etiketiyle (PNEUMONIA=0, NORMAL=1) birlikte kaydeder.<br>
* Tm verileri bir NumPy array olarak geri dndrr.<br>

  Grnt verilerinin derin renme modellerinde ilenebilmesi iin tm verilerin NumPy array formatnda tutulmas gerekmektedir. Bu yntem, verilerin GPU veya CPU zerinde verimli bir ekilde ilenmesini, eitim srasnda batch tabanl veri aknn salanmasn ve tm girilerin boyut ve veri tipi asndan tutarl olmasn garanti eder. Ayrca NumPy array kullanm, modern derin renme ktphaneleri (TensorFlow, Keras, PyTorch) ile tam uyumluluk salayarak, veri n ileme ve normalizasyon ilemlerinin tm veri setine hzl ve kolay bir ekilde uygulanmasna olanak tanr. Bu sayede eitim sreci hem hzl hem de stabil bir ekilde yrtlebilmektedir.

# MODELLEME(CNN)

### Convolutional Neural Network (CNN) Modeli Aklamas

Bu alma kapsamnda, **gs rntgeni (Chest X-Ray) verilerini pnmoni ve normal olarak snflandrmak** iin derin bir Convolutional Neural Network (CNN) kullanlmtr. Modelin mimarisi ve bileenleri aada aklanmtr.

#### 1 Model Katmanlar ve Teknik Terimler

1. **Conv2D (Convolutional Layer)**  
   - rnek: `Conv2D(32, (3,3), strides=1, padding='same', activation='relu', input_shape=(150,150,1))`  
   - Grntden zellik karan temel katmandr.  
   - **32 filtre**, **3x3 ekirdek boyutu**, `stride=1` ile kaydrma yapar.  
   - `padding='same'` ile kt boyutu giri ile ayn tutulur.  
   - `activation='relu'` ile dorusal olmayan (non-linear) aktivasyon uygulanr.

2. **BatchNormalization**  
   - Her mini-batch iin aktivasyonlar normalize eder.  
   - Eitim hzn artrr ve modelin kararlln salar.

3. **MaxPool2D (Max Pooling)**  
   - rnek: `MaxPool2D((2,2), strides=2, padding='same')`  
   - zellik haritalarn kltr (downsampling).  
   - En yksek aktivasyonu alarak boyutu yarya indirir.  
   - Modelin parametre saysn azaltr ve overfitting riskini drr.

4. **Dropout**  
   - rnek: `Dropout(0.2)`  
   - Eitim srasnda belirli oranlarda nronlar rastgele devre d brakr.  
   - Overfittingi azaltmaya yardmc olur.

5. **Flatten**  
   - ok boyutlu (2D) feature mapleri tek boyutlu vektre evirir.  
   - Dense (tam balantl) katmana girdi salar.

6. **Dense (Fully Connected Layer)**  
   - rnek: `Dense(128, activation='relu')`  
   - Tm nronlar nceki katmandaki tm nronlara baldr.  
   - renilmi zellikleri birletirerek snflandrma yapar.

7. **k Katman (Output Layer)**  
   - `Dense(1, activation='sigmoid')`  
   - Binary snflandrma iin tek nron ve sigmoid aktivasyonu kullanlr.  
   - k deeri **0 ile 1** arasnda olup, olaslk olarak yorumlanr.

---

#### 2 Modelin Derlenmesi (Compilation)

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])


````python
# -------------------------------
# CNN Model
# -------------------------------
model = Sequential([
    Conv2D(32, (3,3), strides=1, padding='same', activation='relu', input_shape=(150,150,1)),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),


    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.1),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(128, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(256, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')  # Binary classification
])

# -------------------------------
# Compile
# -------------------------------
model.compile(optimizer = "rmsprop" , loss = 'binary_crossentropy' , metrics = ['accuracy'])

model.summary()
````

````python
# -------------------------------
# Callback
# -------------------------------
learning_rate_reduction = ReduceLROnPlateau(
    monitor='val_accuracy',
    patience=2,
    verbose=1,
    factor=0.3,
    min_lr=0.000001
)

# -------------------------------
# Model Eitimi
# -------------------------------
history = model.fit(
    train_generator,
    epochs=12,
    validation_data=val_generator,
    callbacks=[learning_rate_reduction]
)

# -------------------------------
# Test seti deerlendirme
# -------------------------------
test_datagen = ImageDataGenerator()
test_generator = test_datagen.flow(x_test, y_test, batch_size=32, shuffle=False)

test_loss, test_acc = model.evaluate(test_generator)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc*100:.2f}%")
````

# HEATMAP LE GRSELLETRME(Grad_CAM)

## Grad-CAM: Derin renmede Kullanm ve Yorumlanmas

### Giri
Derin renme modelleri, zellikle Convolutional Neural Networks (CNN), grnt snflandrma, nesne tanma ve tbbi grnt analizi gibi alanlarda son derece baarl sonular vermektedir. Ancak bu modeller, "black-box" olarak adlandrlan yaplar nedeniyle, modelin **hangi blgelere dikkat ederek karar verdiini anlamak zordur**. Bu durum, zellikle tbbi uygulamalarda modelin kararlarnn gvenilirliini deerlendirmek iin kritik bir sorundur.

Grad-CAM (Gradient-weighted Class Activation Mapping), derin renme modellerinin **karar mekanizmasn grselletirmek ve yorumlamak** amacyla gelitirilmi bir yntemdir.

---

### Grad-CAMin Amac ve Kullanm Alanlar

1. **Modelin dikkatini analiz etmek:**  
   Grad-CAM, snflandrma srasnda modelin hangi **piksel veya blgeleri daha fazla dikkate aldn** ortaya karr. Bu sayede modelin kararlar daha effaf hale gelir.

2. **Grselletirme ve aklanabilirlik:**  
   - zellikle tbbi grntleme, otonom aralar ve gvenlik sistemlerinde, modelin **odakland alanlarn grselletirilmesi**, kullancya modelin mantn aklar.  
   - rnein bir gs rntgeninde pnmoni tespiti yapan model, Grad-CAM ile **akcierin hangi blgesine dikkat ettiini** gsterebilir.

3. **Hatal kararlarn analizi:**  
   Grad-CAM sayesinde modelin yanl snflandrd rneklerde **dikkat dalm incelenebilir**. Bu, modelin renme srecindeki eksiklikleri belirlemek iin kullanlr.

---

### Grad-CAMin alma Prensibi

1. **Backpropagation ile gradient hesaplama:**  
   - Grad-CAM, hedef snf iin son konvolsyon katmanndaki feature maplerin **gradientlerini hesaplar**.  
   - Bu gradientler, hangi feature mapin snfa katksnn yksek olduunu gsterir.

2. **Feature map arlklandrmas:**  
   - Hesaplanan gradientler ile feature mapler arplr ve arlkl toplam alnr.  
   - Sonu, **snfa zg dikkat haritas** (class activation map) retir.

3. **Grselletirme:**  
   - Oluan dikkat haritas, orijinal grnt zerinde **heatmap olarak st ste bindirilir**.  
   - Krmz blgeler modelin daha ok dikkat ettii alanlar, mavi blgeler daha az dikkat ettii alanlar gsterir.
4. **Grad-CAM Renklerin Yorumu**
   - Grad-CAM yntemi, derin renme modellerinin karar mekanizmasn grselletirmek amacyla kullanlr. Bu yntemde oluturulan renkli haritalar, modelin **girdi grntsndeki hangi blgelere ne kadar dikkat ettiini** gsterir. Baka bir deyile, her pikselin snf tahminine katksnn bykl renk younluu ile ifade edilir.

Grad-CAM grselletirmelerinde kullanlan renk skalas (rnein **jet colormap**) genellikle **mavi  yeil  sar  krmz** eklindedir:

- **Mavi / souk renkler:** Modelin dikkatinin daha az olduu blgeler.  
- **Yeil / sar renkler:** Modelin orta derecede odakland blgeler.  
- **Krmz / scak renkler:** Modelin snf tahmininde en ok odakland, yani en nemli bulduu blgeler.

Bu younluk dalm, modelin karar srecini yorumlamada kritik bir gstergedir.  
- **Doru snflandrmalarda**, krmz blgeler genellikle ilgili veya hastalkl dokular ile rtr.  
- **Yanl snflandrmalarda**, krmz alanlar modelin alakasz veya hatal blgelere odaklandn ortaya koyar.

zetle, Grad-CAM renk younluu, modelin **ilgisi ve snf tahminine katksn** grselletirmenin gl bir yoludur.


---

## Grad-CAMin Yorumlanmas

- **Krmz blgeler:** Modelin snf tahminine en fazla katkda bulunan alanlar.  
- **Mavi veya dk younluklu blgeler:** Modelin kararnda daha az etkili olan alanlar.  
- **Uygulama rnei:** Bir gs rntgeninde pnmoni tespiti yaplyorsa, krmz blgeler **enflamasyon veya lezyon bulunan akcier blgeleri** ile rtmelidir.  
- **Yorumlama ipular:**  
  1. Modelin dikkat ettii blgeler, domain uzman tarafndan **doru klinik blgelerle karlatrlabilir**.  
  2. Yanl snflandrmalarda, Grad-CAM, modelin **gereksiz veya alakasz blgelere dikkat ettiini** gsterebilir.  
  3. Bylece modelin eitimi veya veri seti zerinde iyiletirme yaplabilir.

---



Grad-CAM, derin renme modellerinin karar srelerini **anlaml ve grselletirilebilir bir ekilde aa karan gl bir yntemdir**. Modelin hangi alanlara dikkat ettiini grselletirmek, zellikle kritik uygulamalarda **gvenilirlik ve aklanabilirlik** salar. Tez almalar ve tbbi grnt analizlerinde Grad-CAM, **model performansn yorumlamak, hatalar analiz etmek ve model gvenini artrmak** iin standart bir ara olarak kullanlmaktadr. Bu sebeple, modelin karar mekanizmasn grselletirmek amacyla heatmap tabanl Grad-CAM metodu seilmitir.


````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model

labels = ["PNEUMONIA", "NORMAL"]  # Snf isimleri

# -----------------------------
# Basit Grad-CAM fonksiyonu
# -----------------------------
def gradcam(model, img, class_idx):
    for layer in reversed(model.layers):
        if 'conv' in layer.name:
            last_conv = layer
            break

    grad_model = Model(model.inputs, [last_conv.output, model.output])
    img_tensor = np.expand_dims(img, axis=0)

    with tf.GradientTape() as tape:
        conv_out, pred = grad_model(img_tensor)
        loss = pred[:, class_idx]

    grads = tape.gradient(loss, conv_out)
    weights = tf.reduce_mean(grads, axis=(0,1,2))
    cam = tf.reduce_sum(weights * conv_out[0], axis=-1)
    cam = np.maximum(cam, 0)
    cam = cam / (np.max(cam) + 1e-8)
    cam = tf.image.resize(cam[..., np.newaxis], (img.shape[0], img.shape[1])).numpy()[:,:,0]
    return cam

# -----------------------------
# Heatmap overlay
# -----------------------------
def overlay(img, heatmap, alpha=0.4):
    if img.ndim == 2:
        img_rgb = np.stack([img]*3, axis=-1)
    else:
        img_rgb = img
    heatmap_rgb = plt.get_cmap('jet')(heatmap)[:,:,:3]
    overlayed = img_rgb + heatmap_rgb * alpha
    overlayed = np.clip(overlayed, 0, 1)
    return overlayed

# -----------------------------
# Test tahminleri
# -----------------------------
preds = (model.predict(x_test) > 0.5).astype(int).flatten()

# lk doru ve yanl tahminleri bul
correct_idx = np.where(preds == y_test)[0][0]
wrong_idx   = np.where(preds != y_test)[0][0]

# Seilen rnekler
img_correct = x_test[correct_idx]
img_wrong   = x_test[wrong_idx]

# Snf indekslerini integer yap
class_correct = int(y_test[correct_idx])
class_wrong   = int(y_test[wrong_idx])
pred_wrong_int = int(preds[wrong_idx])

# Grad-CAM
heat_correct = gradcam(model, img_correct, class_correct)
heat_wrong   = gradcam(model, img_wrong, class_wrong)

# Overlay
overlay_correct = overlay(img_correct, heat_correct)
overlay_wrong   = overlay(img_wrong, heat_wrong)

# Grselletir
fig, axes = plt.subplots(2,3, figsize=(12,6))

axes[0,0].imshow(img_correct.squeeze(), cmap='gray')
axes[0,0].set_title(f"Doru - Orijinal\nSnf: {labels[class_correct]}")
axes[0,0].axis('off')
axes[0,1].imshow(heat_correct, cmap='jet')
axes[0,1].set_title("Grad-CAM")
axes[0,1].axis('off')
axes[0,2].imshow(overlay_correct)
axes[0,2].set_title("Overlay")
axes[0,2].axis('off')

axes[1,0].imshow(img_wrong.squeeze(), cmap='gray')
axes[1,0].set_title(f"Yanl - Orijinal\nGerek: {labels[class_wrong]}\nTahmin: {labels[pred_wrong_int]}")
axes[1,0].axis('off')
axes[1,1].imshow(heat_wrong, cmap='jet')
axes[1,1].set_title("Grad-CAM")
axes[1,1].axis('off')
axes[1,2].imshow(overlay_wrong)
axes[1,2].set_title("Overlay")
axes[1,2].axis('off')

plt.tight_layout()
plt.show()
````

**Gerek Snf:** PNEUMONIA  
**Model Tahmini:** NORMAL  

**Grad-CAM Heatmap:** Krmz ve sar blgeler akcierlerin alt younlamtr.  
**Overlay:** Model, karar verirken bu alanlara odakland iin doru grubu seememi.

#### Olas Nedenler
1. **Yanl odaklanma / dikkat sapmas:** Model, alt akcierlerdeki younluk deiimlerini pnmoni olarak alglamam olabilir.  
2. **Snf karkl:** Pnmoni ve normal akcierler benzer opasite rntlerine sahip olabilir; eitim verisindeki normal rnekler modelin kararn etkileyebilir.  
3. **Veri ve model snrlamalar:** Dk kontrastl veya az belirgin lezyonlar CNN tarafndan zor alglanm olabilir.  

#### Grsel-Model Balants
Grad-CAM krmz blgeleri modelin dikkat ettii alanlar gsterir. Overlay ile bu blgeler orijinal X-ray zerinde grselletirilmitir. Model, kritik blgeleri grmesine ramen zellikleri pnmoni olarak snflandramamtr; bu durum eitim verisi eksiklii veya modelin feature learning snrlar ile ilikilidir.

# Hiperparametre Optimizasyonu

### Veri Setinin Yeninden Yklenmesi

````python
import numpy as np

# Kaydedilen dataset'i ykle
data = np.load("combined_dataset.npz")

# indekileri deikenlere ayr
x_train = data["x_train"]
y_train = data["y_train"]
x_val   = data["x_val"]
y_val   = data["y_val"]
x_test  = data["x_test"]
y_test  = data["y_test"]
````

# MODELLEME (CNN)

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau

# -------------------------------
# Train ve Validation Data Augmentation
# -------------------------------
train_datagen = ImageDataGenerator(
    rotation_range=30,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=True
)

val_datagen = ImageDataGenerator()  # sadece augmentationsz generator

# -------------------------------
# Generator tanmlar
# -------------------------------
train_generator = train_datagen.flow(x_train, y_train, batch_size=32, shuffle=True)
val_generator = val_datagen.flow(x_val, y_val, batch_size=32, shuffle=False)

# -------------------------------
# CNN Model
# -------------------------------
model = Sequential([
    Conv2D(32, (3,3), strides=1, padding='same', activation='relu', input_shape=(150,150,1)),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),


    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.1),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(128, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # Binary classification
])

# -------------------------------
# Compile
# -------------------------------
model.compile(optimizer = "rmsprop" , loss = 'binary_crossentropy' , metrics = ['accuracy'])

model.summary()
````

````python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)

learning_rate_reduction = ReduceLROnPlateau(
    monitor='val_accuracy', 
    patience = 2, 
    verbose=1,
    factor=0.3, 
    min_lr=0.000001
)
````

````python
# -------------------------------
# Model Eitimi
# -------------------------------
history = model.fit(train_generator,epochs = 12 , 
                    validation_data=val_generator,
                    callbacks = [early_stop, learning_rate_reduction])
````
