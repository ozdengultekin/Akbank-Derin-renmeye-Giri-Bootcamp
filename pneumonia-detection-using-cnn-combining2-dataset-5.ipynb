# VERİ ÖN İŞLEME ve GÖRSELLEŞTİRME

**Sınıflar ve Görüntü Boyutu:**<br>
Labels listesi, veri setindeki sınıfları temsil eder: PNEUMONIA ve NORMAL. Tüm görüntüler 150x150 piksel boyutuna getirilir; bu, CNN modellerinde giriş boyutunun sabit olması için gereklidir.<br>
**Ön İşleme:**
**CLAHE (Contrast Limited Adaptive Histogram Equalization)** ile görüntü kontrastı artırılır. Böylece akciğer dokusundaki küçük detaylar ve opasiteler daha belirgin hâle gelir. Normalizasyon uygulanırsa, piksel değerleri 0–1 aralığına ölçeklenir; bu, modelin daha hızlı ve kararlı öğrenmesini sağlar. Son olarak reshape işlemi ile görüntüler CNN giriş formatına (yükseklik, genişlik, kanal) getirilir.<br>
**Veri Setinin Hazırlanması:**
Her resim, uygun preprocessing adımlarından geçtikten sonra kendi sınıf etiketi ile birlikte [resim, etiket] çifti hâline getirilir. Bu sayede veri seti modelin eğitimine hazır hale gelir.<br>
**Özellikler ve Etiketlerin Ayrılması:**
X dizisi modelin girişine verilecek görüntüleri, Y dizisi ise her görüntünün sınıf etiketini içerir. Bu ayrım, modelin giriş ve hedef dizilerini net bir şekilde ayırır.

````python
labels = ["PNEUMONIA", "NORMAL"]
img_size = 150  # 150x150 piksel

#Normalizasyon ve Yeniden Boyutlandırma

def preprocess_image(img):
    # CLAHE ile kontrast artır
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img = clahe.apply(img)
    # Normalize et
    img = img.astype(np.float32) / 255.0
    # CNN için reshape
    img = img.reshape(img_size, img_size, 1)
    return img

def get_training_data(data_dir):
    dataset = []
    for class_id, label in enumerate(labels):
        label_dir = os.path.join(data_dir, label)
        for file in os.listdir(label_dir):
            img_path = os.path.join(label_dir, file)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = cv2.resize(img, (img_size, img_size))
                img = preprocess_image(img)  # Ön işleme + normalize + reshape
                dataset.append([img, class_id])
    return np.array(dataset)

# Özellikler ve etiketleri ayır
def split_features_labels(dataset):
    x = [feature for feature, label in dataset]
    y = [label for feature, label in dataset]
    return np.array(x), np.array(y)
````

**PNEUMONIA** ve **NORMAL** klasöründe bulunan resimleri:

* Grayscale (siyah-beyaz) olarak okur,150x150 boyutuna küçültür. img_size klasörlerde bulunan her resim boyutu farklı. Bu görüntüleri işlerken her bir görüntünün eşit boyutta olması için bu boyutu her resim için aynı boyuta getiriryoruz. Bu ayrıca modelin resimleri daha hızlı işlemesini sağlar. Resim büyüdükçe modelin eğitim süresinin uzadığı da gözlemlenmiştir.
PNEUMONIA 0 , NORMAL 1 olarak listede tanımlandı.
* Her resmi ilgili sınıf etiketiyle (PNEUMONIA=0, NORMAL=1) birlikte kaydeder.<br>
* Tüm verileri bir NumPy array olarak geri döndürür.<br>

  Görüntü verilerinin derin öğrenme modellerinde işlenebilmesi için tüm verilerin NumPy array formatında tutulması gerekmektedir. Bu yöntem, verilerin GPU veya CPU üzerinde verimli bir şekilde işlenmesini, eğitim sırasında batch tabanlı veri akışının sağlanmasını ve tüm girişlerin boyut ve veri tipi açısından tutarlı olmasını garanti eder. Ayrıca NumPy array kullanımı, modern derin öğrenme kütüphaneleri (TensorFlow, Keras, PyTorch) ile tam uyumluluk sağlayarak, veri ön işleme ve normalizasyon işlemlerinin tüm veri setine hızlı ve kolay bir şekilde uygulanmasına olanak tanır. Bu sayede eğitim süreci hem hızlı hem de stabil bir şekilde yürütülebilmektedir.

# MODELLEME(CNN)

### Convolutional Neural Network (CNN) Modeli Açıklaması

Bu çalışma kapsamında, **göğüs röntgeni (Chest X-Ray) verilerini pnömoni ve normal olarak sınıflandırmak** için derin bir Convolutional Neural Network (CNN) kullanılmıştır. Modelin mimarisi ve bileşenleri aşağıda açıklanmıştır.

#### 1️⃣ Model Katmanları ve Teknik Terimler

1. **Conv2D (Convolutional Layer)**  
   - Örnek: `Conv2D(32, (3,3), strides=1, padding='same', activation='relu', input_shape=(150,150,1))`  
   - Görüntüden özellik çıkaran temel katmandır.  
   - **32 filtre**, **3x3 çekirdek boyutu**, `stride=1` ile kaydırma yapar.  
   - `padding='same'` ile çıktı boyutu giriş ile aynı tutulur.  
   - `activation='relu'` ile doğrusal olmayan (non-linear) aktivasyon uygulanır.

2. **BatchNormalization**  
   - Her mini-batch için aktivasyonları normalize eder.  
   - Eğitim hızını artırır ve modelin kararlılığını sağlar.

3. **MaxPool2D (Max Pooling)**  
   - Örnek: `MaxPool2D((2,2), strides=2, padding='same')`  
   - Özellik haritalarını küçültür (downsampling).  
   - En yüksek aktivasyonu alarak boyutu yarıya indirir.  
   - Modelin parametre sayısını azaltır ve overfitting riskini düşürür.

4. **Dropout**  
   - Örnek: `Dropout(0.2)`  
   - Eğitim sırasında belirli oranlarda nöronları rastgele devre dışı bırakır.  
   - Overfitting’i azaltmaya yardımcı olur.

5. **Flatten**  
   - Çok boyutlu (2D) feature map’leri tek boyutlu vektöre çevirir.  
   - Dense (tam bağlantılı) katmana girdi sağlar.

6. **Dense (Fully Connected Layer)**  
   - Örnek: `Dense(128, activation='relu')`  
   - Tüm nöronlar önceki katmandaki tüm nöronlara bağlıdır.  
   - Öğrenilmiş özellikleri birleştirerek sınıflandırma yapar.

7. **Çıkış Katmanı (Output Layer)**  
   - `Dense(1, activation='sigmoid')`  
   - Binary sınıflandırma için tek nöron ve sigmoid aktivasyonu kullanılır.  
   - Çıkış değeri **0 ile 1** arasında olup, olasılık olarak yorumlanır.

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])


````python
# -------------------------------
# CNN Model
# -------------------------------
model = Sequential([
    Conv2D(32, (3,3), strides=1, padding='same', activation='relu', input_shape=(150,150,1)),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),


    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.1),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(128, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(256, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')  # Binary classification
])

# -------------------------------
# Compile
# -------------------------------
model.compile(optimizer = "rmsprop" , loss = 'binary_crossentropy' , metrics = ['accuracy'])

model.summary()
````

# HEATMAP İLE GÖRSELLEŞTİRME(Grad_CAM)

## Grad-CAM: Derin Öğrenmede Kullanımı ve Yorumlanması

### Giriş
Derin öğrenme modelleri, özellikle Convolutional Neural Networks (CNN), görüntü sınıflandırma, nesne tanıma ve tıbbi görüntü analizi gibi alanlarda son derece başarılı sonuçlar vermektedir. Ancak bu modeller, "black-box" olarak adlandırılan yapıları nedeniyle, modelin **hangi bölgelere dikkat ederek karar verdiğini anlamak zordur**. Bu durum, özellikle tıbbi uygulamalarda modelin kararlarının güvenilirliğini değerlendirmek için kritik bir sorundur.

Grad-CAM (Gradient-weighted Class Activation Mapping), derin öğrenme modellerinin **karar mekanizmasını görselleştirmek ve yorumlamak** amacıyla geliştirilmiş bir yöntemdir.

---

### Grad-CAM’in Amacı ve Kullanım Alanları

1. **Modelin dikkatini analiz etmek:**  
   Grad-CAM, sınıflandırma sırasında modelin hangi **piksel veya bölgeleri daha fazla dikkate aldığını** ortaya çıkarır. Bu sayede modelin kararları daha şeffaf hale gelir.

2. **Görselleştirme ve açıklanabilirlik:**  
   - Özellikle tıbbi görüntüleme, otonom araçlar ve güvenlik sistemlerinde, modelin **odaklandığı alanların görselleştirilmesi**, kullanıcıya modelin mantığını açıklar.  
   - Örneğin bir göğüs röntgeninde pnömoni tespiti yapan model, Grad-CAM ile **akciğerin hangi bölgesine dikkat ettiğini** gösterebilir.

3. **Hatalı kararların analizi:**  
   Grad-CAM sayesinde modelin yanlış sınıflandırdığı örneklerde **dikkat dağılımı incelenebilir**. Bu, modelin öğrenme sürecindeki eksiklikleri belirlemek için kullanılır.

---

### Grad-CAM’in Çalışma Prensibi

1. **Backpropagation ile gradient hesaplama:**  
   - Grad-CAM, hedef sınıf için son konvolüsyon katmanındaki feature map’lerin **gradientlerini hesaplar**.  
   - Bu gradientler, hangi feature map’in sınıfa katkısının yüksek olduğunu gösterir.

2. **Feature map ağırlıklandırması:**  
   - Hesaplanan gradientler ile feature map’ler çarpılır ve ağırlıklı toplam alınır.  
   - Sonuç, **sınıfa özgü dikkat haritası** (class activation map) üretir.

3. **Görselleştirme:**  
   - Oluşan dikkat haritası, orijinal görüntü üzerinde **heatmap olarak üst üste bindirilir**.  
   - Kırmızı bölgeler modelin daha çok dikkat ettiği alanları, mavi bölgeler daha az dikkat ettiği alanları gösterir.
4. **Grad-CAM Renklerin Yorumu**
   - Grad-CAM yöntemi, derin öğrenme modellerinin karar mekanizmasını görselleştirmek amacıyla kullanılır. Bu yöntemde oluşturulan renkli haritalar, modelin **girdi görüntüsündeki hangi bölgelere ne kadar dikkat ettiğini** gösterir. Başka bir deyişle, her pikselin sınıf tahminine katkısının büyüklüğü renk yoğunluğu ile ifade edilir.

Grad-CAM görselleştirmelerinde kullanılan renk skalası (örneğin **jet colormap**) genellikle **mavi → yeşil → sarı → kırmızı** şeklindedir:

- **Mavi / soğuk renkler:** Modelin dikkatinin daha az olduğu bölgeler.  
- **Yeşil / sarı renkler:** Modelin orta derecede odaklandığı bölgeler.  
- **Kırmızı / sıcak renkler:** Modelin sınıf tahmininde en çok odaklandığı, yani en önemli bulduğu bölgeler.

Bu yoğunluk dağılımı, modelin karar sürecini yorumlamada kritik bir göstergedir.  
- **Doğru sınıflandırmalarda**, kırmızı bölgeler genellikle ilgili veya hastalıklı dokular ile örtüşür.  
- **Yanlış sınıflandırmalarda**, kırmızı alanlar modelin alakasız veya hatalı bölgelere odaklandığını ortaya koyar.

Özetle, Grad-CAM renk yoğunluğu, modelin **ilgisi ve sınıf tahminine katkısını** görselleştirmenin güçlü bir yoludur.


---

## Grad-CAM’in Yorumlanması

- **Kırmızı bölgeler:** Modelin sınıf tahminine en fazla katkıda bulunan alanlar.  
- **Mavi veya düşük yoğunluklu bölgeler:** Modelin kararında daha az etkili olan alanlar.  
- **Uygulama örneği:** Bir göğüs röntgeninde pnömoni tespiti yapılıyorsa, kırmızı bölgeler **enflamasyon veya lezyon bulunan akciğer bölgeleri** ile örtüşmelidir.  
- **Yorumlama ipuçları:**  
  1. Modelin dikkat ettiği bölgeler, domain uzmanı tarafından **doğru klinik bölgelerle karşılaştırılabilir**.  
  2. Yanlış sınıflandırmalarda, Grad-CAM, modelin **gereksiz veya alakasız bölgelere dikkat ettiğini** gösterebilir.  
  3. Böylece modelin eğitimi veya veri seti üzerinde iyileştirme yapılabilir.

---



Grad-CAM, derin öğrenme modellerinin karar süreçlerini **anlamlı ve görselleştirilebilir bir şekilde açığa çıkaran güçlü bir yöntemdir**. Modelin hangi alanlara dikkat ettiğini görselleştirmek, özellikle kritik uygulamalarda **güvenilirlik ve açıklanabilirlik** sağlar. Tez çalışmaları ve tıbbi görüntü analizlerinde Grad-CAM, **model performansını yorumlamak, hataları analiz etmek ve model güvenini artırmak** için standart bir araç olarak kullanılmaktadır. Bu sebeple, modelin karar mekanizmasını görselleştirmek amacıyla heatmap tabanlı Grad-CAM metodu seçilmiştir.


````python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model

labels = ["PNEUMONIA", "NORMAL"]  # Sınıf isimleri

# -----------------------------
# Basit Grad-CAM fonksiyonu
# -----------------------------
def gradcam(model, img, class_idx):
    for layer in reversed(model.layers):
        if 'conv' in layer.name:
            last_conv = layer
            break

    grad_model = Model(model.inputs, [last_conv.output, model.output])
    img_tensor = np.expand_dims(img, axis=0)

    with tf.GradientTape() as tape:
        conv_out, pred = grad_model(img_tensor)
        loss = pred[:, class_idx]

    grads = tape.gradient(loss, conv_out)
    weights = tf.reduce_mean(grads, axis=(0,1,2))
    cam = tf.reduce_sum(weights * conv_out[0], axis=-1)
    cam = np.maximum(cam, 0)
    cam = cam / (np.max(cam) + 1e-8)
    cam = tf.image.resize(cam[..., np.newaxis], (img.shape[0], img.shape[1])).numpy()[:,:,0]
    return cam

# -----------------------------
# Heatmap overlay
# -----------------------------
def overlay(img, heatmap, alpha=0.4):
    if img.ndim == 2:
        img_rgb = np.stack([img]*3, axis=-1)
    else:
        img_rgb = img
    heatmap_rgb = plt.get_cmap('jet')(heatmap)[:,:,:3]
    overlayed = img_rgb + heatmap_rgb * alpha
    overlayed = np.clip(overlayed, 0, 1)
    return overlayed

# -----------------------------
# Test tahminleri
# -----------------------------
preds = (model.predict(x_test) > 0.5).astype(int).flatten()

# İlk doğru ve yanlış tahminleri bul
correct_idx = np.where(preds == y_test)[0][0]
wrong_idx   = np.where(preds != y_test)[0][0]

# Seçilen örnekler
img_correct = x_test[correct_idx]
img_wrong   = x_test[wrong_idx]

# Sınıf indekslerini integer yap
class_correct = int(y_test[correct_idx])
class_wrong   = int(y_test[wrong_idx])
pred_wrong_int = int(preds[wrong_idx])

# Grad-CAM
heat_correct = gradcam(model, img_correct, class_correct)
heat_wrong   = gradcam(model, img_wrong, class_wrong)

# Overlay
overlay_correct = overlay(img_correct, heat_correct)
overlay_wrong   = overlay(img_wrong, heat_wrong)

# Görselleştir
fig, axes = plt.subplots(2,3, figsize=(12,6))

axes[0,0].imshow(img_correct.squeeze(), cmap='gray')
axes[0,0].set_title(f"Doğru - Orijinal\nSınıf: {labels[class_correct]}")
axes[0,0].axis('off')
axes[0,1].imshow(heat_correct, cmap='jet')
axes[0,1].set_title("Grad-CAM")
axes[0,1].axis('off')
axes[0,2].imshow(overlay_correct)
axes[0,2].set_title("Overlay")
axes[0,2].axis('off')

axes[1,0].imshow(img_wrong.squeeze(), cmap='gray')
axes[1,0].set_title(f"Yanlış - Orijinal\nGerçek: {labels[class_wrong]}\nTahmin: {labels[pred_wrong_int]}")
axes[1,0].axis('off')
axes[1,1].imshow(heat_wrong, cmap='jet')
axes[1,1].set_title("Grad-CAM")
axes[1,1].axis('off')
axes[1,2].imshow(overlay_wrong)
axes[1,2].set_title("Overlay")
axes[1,2].axis('off')

plt.tight_layout()
plt.show()
````

### Grad-CAM Analizi – Hatalı Tahmin Örneği

**Gerçek Sınıf:** PNEUMONIA  
**Model Tahmini:** NORMAL  

**Grad-CAM Heatmap:** Kırmızı ve sarı bölgeler akciğerlerin alt yoğunlaşmıştır.  
**Overlay:** Model, karar verirken bu alanlara odaklandığı için doğru grubu seçememiş.

#### Olası Nedenler
1. **Yanlış odaklanma / dikkat sapması:** Model, alt akciğerlerdeki yoğunluk değişimlerini pnömoni olarak algılamamış olabilir.  
2. **Sınıf karışıklığı:** Pnömoni ve normal akciğerler benzer opasite örüntülerine sahip olabilir; eğitim verisindeki normal örnekler modelin kararını etkileyebilir.  
3. **Veri ve model sınırlamaları:** Düşük kontrastlı veya az belirgin lezyonlar CNN tarafından zor algılanmış olabilir.  

#### Görsel-Model Bağlantısı
Grad-CAM kırmızı bölgeleri modelin dikkat ettiği alanları gösterir. Overlay ile bu bölgeler orijinal X-ray üzerinde görselleştirilmiştir. Model, kritik bölgeleri görmesine rağmen özellikleri pnömoni olarak sınıflandıramamıştır; bu durum eğitim verisi eksikliği veya modelin feature learning sınırları ile ilişkilidir.

#### İyileştirme Önerileri
- Eğitim verisine düşük kontrastlı pnömoni örnekleri eklemek.  
- Data augmentation (döndürme, parlaklık/kontrast değişimi) ile modelin genel özellik öğrenimini artırmak.  
- Daha derin veya residual bağlantılı mimariler kullanmak.  
- Attention mekanizmaları ekleyerek kritik bölgelere odaklanmayı iyileştirmek.

**Özet:**  
Model, önemli bölgeleri kısmen doğru belirlemiş (Grad-CAM ile gözlemleniyor), fakat sınıflandırmada başarısız olmuştur. 

# Hiperparametre Optimizasyonu

### Veri Setinin Yeninden Yüklenmesi

````python
import numpy as np

# Kaydedilen dataset'i yükle
data = np.load("combined_dataset.npz")

# İçindekileri değişkenlere ayır
x_train = data["x_train"]
y_train = data["y_train"]
x_val   = data["x_val"]
y_val   = data["y_val"]
x_test  = data["x_test"]
y_test  = data["y_test"]
````

# MODELLEME (CNN)

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau

# -------------------------------
# Train ve Validation Data Augmentation
# -------------------------------
train_datagen = ImageDataGenerator(
    rotation_range=30,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=True
)

val_datagen = ImageDataGenerator()  # sadece augmentationsız generator

# -------------------------------
# Generator tanımları
# -------------------------------
train_generator = train_datagen.flow(x_train, y_train, batch_size=32, shuffle=True)
val_generator = val_datagen.flow(x_val, y_val, batch_size=32, shuffle=False)

# -------------------------------
# CNN Model
# -------------------------------
model = Sequential([
    Conv2D(32, (3,3), strides=1, padding='same', activation='relu', input_shape=(150,150,1)),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),


    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.1),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(64, (3,3), strides=1, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Conv2D(128, (3,3), strides=1, padding='same', activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    MaxPool2D((2,2), strides=2, padding='same'),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')  # Binary classification
])

# -------------------------------
# Compile
# -------------------------------
model.compile(optimizer = "rmsprop" , loss = 'binary_crossentropy' , metrics = ['accuracy'])

model.summary()
````

````python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=2,
    restore_best_weights=True
)

learning_rate_reduction = ReduceLROnPlateau(
    monitor='val_accuracy', 
    patience = 2, 
    verbose=1,
    factor=0.3, 
    min_lr=0.000001
)
````

````python
# -------------------------------
# Model Eğitimi
# -------------------------------
history = model.fit(train_generator,epochs = 12 , 
                    validation_data=val_generator,
                    callbacks = [early_stop, learning_rate_reduction])
````

## 📈 Model Eğitim ve Doğrulama Sonuçları (CNN)

### Modelde Yapılan Değişiklikler
Bu bölümde modelin 12 epoch boyunca eğitim ve doğrulama performansı incelenmiştir. 
* Bir önceki modelden farklı olarak sondaki katman azaltıldı.
* **early_stop** eklenmiştir. Early stopping, derin öğrenme eğitiminde sık kullanılan bir regularization (düzenleme) tekniğidir ve temel amacı modelin overfitting yapmasını önlemektir.
* Düzleştirme katmanında dropout değeri 0.3 e yükseltilmiştir. Bunun amacı,  Modelin belli nöronlara aşırı bağımlı olmasını önlemek ve genelleme yeteneğini artırmak. 

````python
# -------------------------------
# Test seti değerlendirme
# -------------------------------

test_datagen = ImageDataGenerator()
test_generator = test_datagen.flow(x_test, y_test, batch_size=32, shuffle=False)

test_loss, test_acc = model.evaluate(test_generator)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc*100:.2f}%")
````
